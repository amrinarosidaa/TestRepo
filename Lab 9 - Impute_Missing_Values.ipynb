```python
import pandas as pd # Jawaban Pertanyaan #1

# Load dataset (Jawaban Pertanyaan #2)
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

print("Dataframe berhasil dimuat.")
print(df.head())
```

```python
# Task 1 & 2: Identifikasi dan hapus duplikat (Jawaban Pertanyaan #3)
duplicate_count = df.duplicated().sum()
print(f"Jumlah baris duplikat ditemukan: {duplicate_count}")

df.drop_duplicates(inplace=True)
print("Baris duplikat telah dihapus.")
```

```python
# Task 3: Missing values untuk semua kolom (Jawaban Pertanyaan #4)
print("\nJumlah nilai yang hilang per kolom:")
print(df.isnull().sum())

# Task 4: Missing values di kolom RemoteWork (Jawaban Pertanyaan #5)
missing_remote = df['RemoteWork'].isnull().sum()
print(f"\nJumlah baris yang hilang di kolom 'RemoteWork': {missing_remote}")
```

```python
# Task 5: Lihat distribusi nilai di RemoteWork
print("\nDistribusi nilai RemoteWork:")
print(df['RemoteWork'].value_counts())

# Task 6: Cari nilai mayoritas (Mode)
majority_value = df['RemoteWork'].value_counts().idxmax()
print(f"Nilai mayoritas adalah: {majority_value}")

# Task 7: Imputasi/Isi nilai yang kosong (Jawaban Pertanyaan #6)
df['RemoteWork'] = df['RemoteWork'].fillna(majority_value)

# Verifikasi
print(f"Jumlah nilai kosong di RemoteWork setelah imputasi: {df['RemoteWork'].isnull().sum()}")
```
