```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Pertanyaan #1: Mengunduh konten halaman web
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data = requests.get(url).text
```

```python
# Pertanyaan #2: Membuat objek Soup
soup = BeautifulSoup(data, "html.parser")
```

```python
# Pertanyaan #3: Mencari nama bahasa pemrograman dan gaji
# Kita akan mencari tabel dan mengambil datanya
# Cek apakah tabel ditemukan
languages_list = []
table = soup.find('table')

for row in table.find_all('tr'):
    cols = row.find_all('td')
    
    # Berdasarkan struktur tabel di URL tersebut:
    # Kolom 1 (index 1) adalah Language
    # Kolom 3 (index 3) adalah Average Annual Salary
    language_name = cols[1].getText()
    average_salary = cols[3].getText()
    
    # Filter agar header tabel tidak ikut masuk ke list data
    if language_name != "Language":
        languages_list.append([language_name, average_salary])
        print(f"{language_name} ---> {average_salary}")
```

```python
# 4. Menyimpan ke popular-languages.csv (Jawaban Pertanyaan #4)
with open("popular-languages.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Language", "Average Annual Salary"]) # Menulis Header
    writer.writerows(languages_list) # Menulis semua data

print("\nSukses! File 'popular-languages.csv' telah dibuat.")
```
