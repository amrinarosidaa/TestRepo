```python
# Load Dataset (Menjawab Pertanyaan #2)

import pandas as pd

# Load dataset (Jawaban Pertanyaan #2)
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

print(f"Data berhasil dimuat. Jumlah baris awal: {df.shape[0]}")
```

```python
# Menghapus Duplikat (Menjawab Pertanyaan #1 & #3)

# Menghitung duplikat sebelum dihapus
print(f"Jumlah duplikat awal: {df.duplicated().sum()}")

# Menghapus duplikat (Jawaban Pertanyaan #3)
df.drop_duplicates(inplace=True)

# Menghapus berdasarkan subset kolom kritis untuk memastikan keunikan (Jawaban Pertanyaan #1)
# Kita asumsikan kolom 'Respondent' (jika ada) atau kombinasi kolom utama adalah kunci unik
df.drop_duplicates(subset=None, keep='first', inplace=True)

print(f"Jumlah baris setelah duplikat dihapus: {df.shape[0]}")
print(f"Verifikasi duplikat setelah dihapus: {df.duplicated().sum()}")
```

```python
# Mengidentifikasi Nilai yang Hilang (Menjawab Pertanyaan #4)

# Cara yang lebih modern untuk mengisi nilai kosong (Imputasi)
most_frequent = df['EdLevel'].value_counts().idxmax()
df['EdLevel'] = df['EdLevel'].fillna(most_frequent) # Menghapus inplace=True

print(f"Nilai kosong di 'EdLevel' telah diisi dengan: {most_frequent}")
```

```python
# Normalisasi Data Kompensasi (Menjawab Pertanyaan #5)
# Mengecek nilai kosong di kolom normalisasi (Jawaban Pertanyaan #5)
missing_comp = df['ConvertedCompYearly'].isnull().sum()
print(f"Jumlah nilai kosong di ConvertedCompYearly: {missing_comp}")

# Mengisi nilai kosong dengan median (metode yang disarankan untuk data gaji)
median_comp = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_comp)

print(f"Normalisasi selesai! Nilai kosong diisi dengan Median: {median_comp}")
```
